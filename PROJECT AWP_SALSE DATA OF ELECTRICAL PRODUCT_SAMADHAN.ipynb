## PROJECT NAME : SALSE DATA OF ELECTRICAL PRODUCT

The following data analysis is based on sales of electrical house products listed in the following columns: 

1] Invoice_ID - Invoice number of sale Product.

2] Product_Code - Unique product code of Product.

3] Product_Name - Name of Product.

4] Sales_Quantity - Count of sale quantity.

5] Unit - Unit of sale quantity.

6] Price - MRP of Product.

7] Currency_Unit - Unit of Price.

8] Product_Cost - Sale price of Product.

9] Currency_Unit2 - Unit of Product cost.

10] Discount - A discount is a reduction in the price of a product or service

11] Company_Name - Product Company name.

12] Country - County of Company.

13] City - City of Company 

14] Sales_Representative - Sales representative is person, who sells products or services for a company and represents their brand.

15] Sale_Date - Product sale date.

16] Delivery Date - Product delivery date.

```python
import pandas as pd
import numpy as mp
import matplotlib.pyplot as plt
import seaborn as sns
```

### Question 1 : What is the count of products represented by each sales representative?

```python
df["Sales_Representative"].value_counts()
```

### Question 2 : What is the maximum discount received for the product?

```python
df["Discount "].max()
```

### Question 3 : What is the maximum product cost of the products that received a 0.15 discount?

```python
discount = df[df['Discount ']==0.15]
max_pcost = discount[' Product_Cost '].max()
max_pcost
```

### Question 4 : What is the minimum price of the product from the company named Bali Company?

```python
company = df[df['Company_Name']=='Bali Company']
min_price = company[' Price '].min()
min_price
```

### Question 5 : What is the maximum sale quantity of the product "dishwasher" sold by Aaron Beatrice?

```python
max_sq = df[(df['Product_Name'] == 'Dish Washer') & (df['Sales_Representative'] == 'Aaron Beatrice')]
max_sq[' Sales_Quantity '].max()
```

### Question 6 : What is the average discount on laptops given by Michael North?

```python
avg_dis = df[(df['Product_Name'] == 'Laptop') & (df['Sales_Representative'] == 'Michael North')]
avg_dis['Discount '].mean()
```

### Question 7 : Which sales representative, invoice ID, and product code correspond to the highest amount of products sold?

```python
df[['Invoice_ID', 'Product_Code','Sales_Representative']].loc[df[' Product_Cost ']==df[' Product_Cost '].max()]
```

### Question 8 : What are the various discounts provided by the sales representatives to customers?

```python
discounts= df['Discount '].unique()
discounts
```

### Question 9 : What percentage of products does the Kapa company have?

```python
KC_Com=df[df['Company_Name']=='Kapa Company'].shape[0]
total_count=df.shape[0]
perct_pd = KC_Com/total_count * 100
perct_pd
```

### Question 10 : What is the Average Sales Quantity with respect to each product?

```python
avg_rate = df.groupby('Product_Name')[' Sales_Quantity '].mean()
avg_rate
```

### Question 11 : Display the price distribution of product price range in graphical terms.?
